```python
year = '2013'
x_ax = 20
_trials = 10
_tresh = 0.65
_filestart = year + 'Attack100/' + year +'-350'
reliability65 = [ ]
_zzz = np.load ( _filestart + '-maxz.npy' )
reliability65.append ( _zzz.flatten ( ) )
for j in range ( 1, x_ax ):
    zzz = [ ]
    for i in range ( _trials ):
        _fileend = '-maxz-' + str ( i ) + '.npy'
        _zzz = np.load ( _filestart + '-' + str ( j ) + _fileend )
        zzz.append ( _zzz.flatten ( ) )
    reliability65.append ( zzz )
_x = _filestart + '-x.npy'
_xx = np.load ( _x )
_y = _filestart + '-y.npy'
_yy = np.load ( _y )
x_flat = _xx.flatten ( )
y_flat = _yy.flatten ( )
m = Basemap ( projection = 'merc', llcrnrlat = 5.5, urcrnrlat = 20.5, llcrnrlon = 115, urcrnrlon = 130, resolution = 'l' )
rel65_2013 = [ 0 ] + [ [ 0 for i in range ( _trials ) ] for j in range ( x_ax - 1 ) ]
for j in range ( len ( x_flat ) ):
    if ( m.is_land ( * m ( x_flat [ j ], y_flat [ j ] ) ) ):
        if ( reliability65 [ 0 ] [ j ] >= _tresh ):
            rel65_2013 [ 0 ] += 1
        for i in range ( 1, x_ax ):
            for k in range ( _trials ):
                if ( reliability65 [ i ] [ k ] [ j ] >= _tresh ) :
                    rel65_2013 [ i ] [ k ] += 1
rel65_2013y = [ rel65_2013 [ 0 ] / _land ] + [ sum ( rel65_2013 [ i ] ) / _trials / _land for i in range ( 1, x_ax ) ]
```
